```python
import pandas as pd
import numpy as np
import pickle as pkl
```

```python
# Read in file of transporter genes
filepath = '../Transporters_TCDB.txt'

tp_df = pd.read_csv(filepath,sep=None,engine='python',header=None)
```

```python
# Read in file of enzyme genes
# filepath = '../Enzyme_genes.txt'
filepath = '../Enzyme_genes2.xlsx'

en_df = pd.read_excel(filepath,sep=None)
```

```python
en_df
```

```python
en_gene_names = en_df['Gene names'].tolist()
```

```python
nan_cnt = 0
en_gene_list = []

for genes in en_gene_names:
    try:
        en_gene_list.extend(genes.split())
    except:
        nan_cnt += 1
```

```python
en_names= np.unique(en_gene_list)
```

```python
tp_df.columns = ['Name','Description','Misc']
```

```python
tp_names = tp_df['Name'].unique().tolist()
```

```python
combined_names = list(np.unique(tp_names + list(en_names)))
```

```python
pkl.dump(combined_names, open( "../combined_names.pkl", "wb" ) )
```

```python
ens_table = pd.read_csv('../Ensembl_symbol_entrez.csv')
```

```python
ens_table
```

```python
combined_names_df = pd.DataFrame(combined_names)
combined_names_df.columns = ['external_gene_name']
```

```python
combined_names_df
```

```python
names_df = pd.merge(ens_table,combined_names_df)
```

```python
names_df.to_csv('../ensembl_index.csv')
```
